The data was obtained from [BigQuery](https://console.cloud.google.com/bigquery) via a query below:

```
SELECT
  year,
  SUM(number) AS number_of_caitlins
FROM
  `bigquery-public-data.usa_names.usa_1910_current`
WHERE
  name = "Caitlin"
GROUP BY
  year
ORDER BY
  year DESC;
```

````
import pandas as pd
````

````
num_caitlins = pd.read_csv('caitlins.csv')
num_caitlins
````

````
num_caitlins_r=num_caitlins[::-1].copy()
num_caitlins_r
````

````
num_theos = pd.read_csv('theos.csv')
````

````
num_katias = pd.read_csv('katias.csv')
````

````
num_helens = pd.read_csv('helens.csv')
````

````
t_a_c = num_theos.merge(num_caitlins_r, on='year')
t_a_c
````

````
k_a_h = num_katias.merge(num_helens, on='year')
k_a_h
````

````
all_together = t_a_c.merge(k_a_h, on='year')
all_together
````

````
all_together.set_index(keys = ["year"], inplace = True)
````

````
all_together.plot(y=["number_of_theos", "number_of_caitlins", "number_of_Katias", "number_of_Helens"], figsize = (10,5), marker = 'o', color = ['b','r','g','c']);
````
